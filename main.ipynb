```python
PATH = ChromeDriverManager().install()    # instala driver de chrome

driver = webdriver.Chrome(PATH)      # abre una ventana de chrome

#Busqueda didn´t work, I had to type up the whole URL + search

URL='https://www.booking.com/searchresults.es.html?ss=Chamart%C3%ADn&ssne=Chamart%C3%ADn&ssne_untouched=Chamart%C3%ADn&efdco=1&label=gen173nr-1BCBkoggI46AdIM1gEaEaIAQGYAQq4ARfIAQzYAQHoAQGIAgGoAgO4AvPn46IGwAIB0gIkOTFmYzYxYTAtYjc3NC00NDM5LWIwNTgtOWIyNGIwYjRiZDA52AIF4AIB&sid=d4b5462ce9ae6cf55f51b42829c8b5ed&aid=304142&lang=es&sb=1&src_elem=sb&src=searchresults&dest_id=180&dest_type=district&checkin=2023-05-08&checkout=2023-05-09&group_adults=2&no_rooms=1&group_children=0&sb_travel_purpose=leisure&flex_window=1'

driver.get(URL)

time.sleep(3)

driver.find_element(By.XPATH, '//*[@id="onetrust-reject-all-handler"]').click()  # rechaza cookies
```

```python
tabla = driver.find_elements(By.CLASS_NAME, 'd20f4628d0')

filas = [e.text.split('\n') for e in tabla]

filas

```

```python
filas_dict = {} #Creates a dictionary with name of accomodation and price.

for i in filas:
    name = i[0].replace('-', '')
    price = i[-3][-3:].strip('.')
    filas_dict[name] = price

filas_dict
```

```python
# Column names
cabeceras = ['name', 'price']

# Creates DF from dictionary
df_booking = pd.DataFrame.from_dict(filas_dict, orient='index')
df_booking = df_booking.reset_index()
df_booking.columns = cabeceras
df_booking = df_booking.rename(columns={'index': 'acc_id'})

df_booking
```
